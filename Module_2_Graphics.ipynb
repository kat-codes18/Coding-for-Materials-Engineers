## Line Plots
The plot function generates a line plot. For example:

    import numpy as np

    x = np.arange(-10,11)

    plt.plot(x,y)

We can add labels for axis, titles and legend with specific details. You should review the choices of fonts, symbols, line styles and colors [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).

    plt.xlabel('X Values')

    plt.legend('Y')

We can position the legend so that it does not detract from the plot adding loc= to teh plt.legend statement. For example:
    plt.legend('Y',loc='upper center')

If needed we can specify the limits of either axis to improve the plot appearance. For example we could use thefollowing for the x axis or similar for y axis:
    plt.xlim(start,end)

When you execute the plot in colab, you will see a reference call in text above the plot. If you wish to avoid this you can type plt.show() as the last line. (This is totally optional and not required)

## A Better Way -Using Subplots

It is often more flexible to use subplots in the plotting routine because this give you more options in what to do with different plots. In Python the **ax** command will allow you to define axs anywhere on the plot screen and allow you to control the position and size of the output plot.

 axes_intro-1.png

 To plot a single figure we open a figure by assigning the plt.subplot to fig,ax (line 6 below). in the subplot function we can specify the plot location and size. Below is the code for a single plot of default size, with the two lines on one plot.

    import matplotlib.pyplot as plt
    import numpy as np
    x = np.linspace(0,2*np.pi,200)
    f1 = np.sin(x)
    f2 = np.cos(x)
    fig, ax = plt.subplots(figsize=())
    ax.plot(x, f1, label='sin(x)')
    ax.plot(x, f2, label='cos(x)')
    lines = ax.plot(x, f1)
    lines= ax.plot(x, f2)
    ax.legend(  loc='lower left')
    plt.show()

To separate the lines, we can use two subplots by changing the assignment to fig,ax .
If we want 2 vertically stacked plots of 4x3 shape, we can specify this by changing the line to:

    fig, axs = plt.subplots(2,figsize=(4,3))

and telling the program where we want each line plotted (which axs)

    axs[0].plot(x, f1)
    axs[1].plot(x, f2)

For horizontal plots we simply change the ax position to (1,2) (meaning 1 row, 2 columns). Note however, the figsize refers to the entire layout you need to adjust the width of the figsize.

    fig, axs = plt.subplots(1,2,figsize=(12,3))

To plot on 2 rows in quadrant 1 and 4 we use:

    fig, axs = plt.subplots(2,2,figsize=(12,6))
    axs[0,0].plot(x, f1)
    axs[1,1].plot(x, f2)

You can plot this data on axes with logarithmically scaled y and linear x using the **plt.semilogy** function, which accepts the same types of arguments as plot. Below, use semilogy to visualize frac_sites vs. T. If you'd like, you can play with some of the other visual tweaks like line color, width, or style. Make sure to label your axes!




label axis and write reason for scaling




```python
import numpy as np
import matplotlib.pyplot as plt
E_D = 0.5
k_B = 8.617e-5
T = np.linspace(100, 500, 100)
N_0 = 1
frac_sites = N_0 * np.exp(E_D / (k_B * T))
fig, axs = plt.subplots(1,2, figsize=(12,3))
axs[0].plot(T, frac_sites)
axs[0].set_xlabel('Temperature (K)')
axs[0].set_ylabel('Fraction of Occupied Defect Sites')
axs[0].set_title('Linear Scale')

axs[1].semilogy(T, frac_sites)
axs[1].set_xlabel('Temperature (K)')
axs[1].set_ylabel('Fraction of Occupied Defect Sites')
axs[1].set_title('Semilogarithmic Scale')
plt.show()
```

## Visualizing Distributions
We can visualize the distribution of a single variable using a histogram with code like:

    plt.hist(data, bins='auto')
    plt.xlabel('Yield Strength (MPa)')
    plt.title('Histogram')

However, histograms require us to choose bins, which can create visual artifacts. Also, it's difficult to compare
two histograms on the same axes to each other (for example, to look at distributions of a property across
materials with different categorical labels). A "swarm chart" can address these two issues, at the expense of
being a bit more qualitative. Look up the documentation for the swarmchart function, and use it below to
generate a plot of the distributions of yield stress for each crystal structure (BCCFCCOther column in the table).
Don't forget to label your axes! In addition, use the title function to give a descriptive title to the plot in the
form of a sentence that describes a trend you observe.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#from google.colab import files
#uploaded = files.upload()

df = pd.read_csv('MPEA - MPEA_dataset-1 (2).csv')


Crystal_Structure = df['PROPERTY: BCC/FCC/other']
yield_stress = df['PROPERTY: YS (MPa)']

plt.figure(figsize=(8, 5))
sns.swarmplot(x='PROPERTY: BCC/FCC/other', y='PROPERTY: YS (MPa)', data=df)

# Label axes and add title
plt.xlabel('Crystal Structure')
plt.ylabel('Yield Stress (MPa)')
plt.title('FCC vs BCC')

# Show plot
plt.show()
```

However, our screens are (for now at least) two-dimensional, so an additional choice we have to make with 3D
visualizations is the angle from which to look at our 3D axes.
We also need  the mplot3d function from the mpl_toolkits library. For plotting lines in 3D we will have to initialize three variable points for the line equation. In our case, we will define three variables as x, y, and z.  

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

# syntax for 3-D projection
ax = plt.axes(projection ='3d')

# defining all 3 axis
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot coding from geeks for geeks')
plt.show()
```
